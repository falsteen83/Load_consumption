```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
from scipy.spatial import cKDTree
import numpy as np
```

```python

# Specify the path to your CSV files
combined_csv_path = "/Users/computer/Downloads/Load/Load/data/combined/*.csv"

# Get all CSV file paths
combined_csv_files = glob.glob(combined_csv_path)

# Read and combine all files into one DataFrame
combined_df = [pd.read_csv(file, parse_dates=['date_time']) for file in combined_csv_files ]

# Merge dataframes of all processed files
combined_df = pd.concat(combined_df, ignore_index=True)

# Display the combined DataFrame
print(combined_df)

```

```python
import pandas as pd

# Assuming combined_df is your DataFrame
missing_values = combined_df.isnull().sum()  # Count missing values for each column
total_values = len(combined_df)  # Total number of rows
missing_percentage = (missing_values / total_values) * 100  # Calculate percentage of missing values

# Create a DataFrame to display the results
missing_info = pd.DataFrame({
    'Column Name': combined_df.columns,
    'Missing Values Count': missing_values,
    'Missing Percentage': missing_percentage
})

# Display the results
print(missing_info)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = combined_df.isnull().mean() * 100

# Get the columns with missing values and their percentages
missing_columns_with_percentage = missing_percentage[missing_percentage > 0]

# Convert to a DataFrame for better display
missing_columns_df = missing_columns_with_percentage.reset_index()
missing_columns_df.columns = ['Column Name', 'Missing Percentage']

print(missing_columns_df)
```

```python
# Select numeric columns from combined_df
numeric_columns = combined_df.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values with the median for each numeric column
combined_df[numeric_columns] = combined_df[numeric_columns].fillna(combined_df[numeric_columns].median())

# Verify that there are no more missing values
missing_values_count = combined_df.isnull().sum()
print(missing_values_count)
```

## Feature selection using correlation

```python
# Selecting only numerical columns from combined_df
numerical_columns = combined_df.select_dtypes(include=['int64', 'float64'])

# Calculating basic statistics for numerical variables
numerical_summary = numerical_columns.describe()
numerical_summary
```

```python
import pandas as pd

# Assuming merged_df is your DataFrame with numerical columns
correlation_matrix = numerical_columns.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the heatmap
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

# Show the plot
plt.title('Correlation Matrix')
plt.show()
```

### step 1

```python
# Identify redundant features (highly correlated pairs)
threshold = 0.8
highly_correlated_pairs = [(i, j) for i in correlation_matrix.columns for j in correlation_matrix.columns 
                           if i != j and abs(correlation_matrix[i][j]) > threshold]

print("Highly correlated pairs (|correlation| > 0.8):")
for pair in highly_correlated_pairs:
    print(pair)
```

```python
# Assess feature correlation with the target variable
target_variable = 'power'  # Replace with your target variable name
correlation_with_target = correlation_matrix[target_variable].sort_values(ascending=False)
print("\nCorrelation with target variable:")
print(correlation_with_target)

# Remove irrelevant features (low correlation with target)
low_correlation_threshold = 0.1  # Adjust this threshold as needed
relevant_features = correlation_with_target[abs(correlation_with_target) >= low_correlation_threshold].index.tolist()

print("\nRelevant features (|correlation| >= 0.1):", relevant_features)

# Create a filtered correlation matrix for relevant features
filtered_correlation_matrix = correlation_matrix.loc[relevant_features, relevant_features]

# Visualize the filtered correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Filtered Correlation Matrix with Target Variable')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming correlation_with_target is already calculated
plt.figure(figsize=(12, 6))

# Create the bar plot without the palette and set colors manually
for i, value in enumerate(correlation_with_target.values):
    color = 'blue' if value >= 0 else 'red'  # Define color based on correlation value
    plt.bar(correlation_with_target.index[i], value, color=color)

plt.title('Correlation with Target Variable: Power')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)  # Rotate feature names for better visibility
plt.axhline(y=0, color='black', linestyle='--')  # Optional: Add a line at y=0 for reference
plt.tight_layout()  # Adjust layout for better spacing
plt.show()

```

# Select feture usingl linear Regression Model

## Step 1: Ensure Your Data is Ready

```python
import pandas as pd
import numpy as np

# Select only numerical columns
X =combined_df.select_dtypes(include=[np.number])  # Keeps only numerical columns

# Define target variable (assuming 'power' is the target)
y = combined_df['power']

# Ensure 'power' is not mistakenly included in features
X = X.drop(columns=['power'], errors='ignore')

# Print selected numerical features
print("Numerical features used for RFE:", X.columns.tolist())
```

## Step 2: Apply Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Initialize the model (you can use other models like RandomForest as well)
model = LinearRegression()

# Use RFE to rank all features
selector = RFE(model, n_features_to_select=1)  # Setting 1 to rank all features
selector.fit(X, y)

# Create a ranking of features
feature_ranking = pd.DataFrame({'Feature': X.columns, 'Ranking': selector.ranking_})
feature_ranking = feature_ranking.sort_values(by='Ranking')

# Display the ranking
print(feature_ranking)

```

##  Step 3: Select Top N Features

```python
n = 5  # Choose the number of top features you want
selector = RFE(model, n_features_to_select=n)
selector.fit(X, y)

# Get selected features
selected_features = X.columns[selector.support_]
print("\nTop", n, "Selected Features:", selected_features.tolist())
```

##  Lasso Regression for Feature Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LassoCV

# Step 1: Ensure Features are Numerical
X = combined_df.select_dtypes(include=[np.number])  
y = combined_df['power']

# Remove target variable if mistakenly included
X = X.drop(columns=['power'], errors='ignore')

# Step 2: Apply Lasso Regression for Feature Selection

lasso = LassoCV(alphas=np.logspace(-3, 3, 100), cv=5, max_iter=5000)
lasso.fit(X, y)

# Get selected features (non-zero coefficients)
selected_features = X.columns[lasso.coef_ != 0]

# Print selected features
print("\nSelected Features using Lasso:", selected_features.tolist())

# Create a DataFrame for feature importance
feature_importance = pd.DataFrame({'Feature': X.columns, 'Coefficient': lasso.coef_})
feature_importance = feature_importance.sort_values(by='Coefficient', ascending=False)

# Step 3: Visualize Feature Importance

plt.figure(figsize=(10, 6))
colors = ['red' if coef < 0 else 'blue' for coef in feature_importance['Coefficient']]

sns.barplot(x=feature_importance['Coefficient'], y=feature_importance['Feature'], 
            palette=colors, legend=False)  # ✅ Explicitly disable `hue`

plt.xlabel('Coefficient Value')
plt.ylabel('Feature Name')
plt.title('Lasso Feature Importance')
plt.axvline(x=0, color='black', linestyle='--')  # Separating positive & negative impact
plt.show()

```

## Random Forst Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Step 1: Ensure Features are Numerical
X = combined_df.select_dtypes(include=[np.number])
y =combined_df['power']

# Remove target variable if mistakenly included
X = X.drop(columns=['power'], errors='ignore')

# Standardize Features (optional but recommended for better performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Train a Random Forest Model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_scaled, y)

# Get Feature Importances
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# Step 3: Print and Visualize Feature Importance
print("\nSelected Features using Random Forest:", feature_importance['Feature'].tolist())

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance['Importance'], y=feature_importance['Feature'], 
            palette='viridis')
plt.xlabel('Feature Importance Score')
plt.ylabel('Feature Name')
plt.title('Feature Importance from Random Forest')
plt.show()
```
